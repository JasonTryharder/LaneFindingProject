```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import math
%matplotlib inline
```

```python
def imageInfo(img):
    X_size = img.shape[1]
    Y_size = img.shape[0]
    print("image type is: ",type(img),"\n Xsize is: ",X_size,"\n Ysize is: ",Y_size)
```

```python
# sobel/gaussian/manual convolution

from scipy import ndimage

# Create a black image
img=np.zeros((640,480))
# ... and make a white rectangle in it
img[100:-100,80:-80]=1

# See how it looks
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
imageInfo(img)
# y = img.shape[0]
# x = img.shape[1]
# print("X size is ",x,"\nY size is ",y)y = img.shape[0]
# x = img.shape[1]
# print("X size is ",x,"\nY size is ",y)
```

```python
# Rotate it for extra fun
img=ndimage.rotate(img,25,mode='constant')
# Have another look
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
imageInfo(img)
```

```python
# Use scipy.ndimage.sobel
# digging deeper into convolution correlation
# Get x-gradient in "sx"
sx = ndimage.sobel(img,axis=0,mode='constant')
# Get y-gradient in "sy"
sy = ndimage.sobel(img,axis=1,mode='constant')
# Get square root of sum of squares
sobel=np.hypot(sx,sy)
plt.imshow(sobel,cmap=plt.cm.gray)
plt.show()

# Use scipy.ndimage and write convolution kernel
# Define kernel for x differences
kx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
# Define kernel for y differences
ky = np.array([[1,2,1] ,[0,0,0], [-1,-2,-1]])
# Perform x convolution
x=ndimage.convolve(img,kx)
# Perform y convolution
y=ndimage.convolve(img,ky)
sobel_convol=np.hypot(x,y)
plt.imshow(sobel,cmap=plt.cm.gray)
plt.show()


# Let's see some edges
```
